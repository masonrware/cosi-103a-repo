### 2. My Solution

```python
import pandas as pd

data = pd.read_csv('./names2k.csv')
namept = pd.pivot_table(data,index='year',values='count',columns='name')
namept[['Emily', 'Hannah', 'Madison', 'Olivia', 'Emma', 'Ava']].plot(title='Popularity of 5 Girls\n Names Over the Last Decade')
```

```python
import numpy as np

print(f'TOP NAMES OF MILLENIA:\n')
data.groupby(['name', 'sex'])['count'].agg(np.sum).sort_values(ascending=False).head(10)
```
